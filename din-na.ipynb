```python
%autoreload 2
import pandas as pd
from os import path
import glob
import warnings
from scrapers.scrapepool import ScrapePool
from scrapers.dinna import DinNaScraper
```

```python
# scraper settings
args = {
    'phantomjs_exec': 'C:\\phantomjs.exe',
    'urls': {
        'save_to': 'data\\din-NA\\out_dec18\\'
    }
}
```

```python
nas = pd.read_csv("data/din-NA/din-NA-toplevel.csv", sep=",", quotechar="\"", encoding="UTF-8").set_index("id")
nas["short"] = nas["name"].apply(lambda n: n.split("(")[-1].split(")")[0])
nas
```

```python
args["nas"] = nas
keywords = nas.index.tolist()
s = ScrapePool(DinNaScraper, keywords, args, chunk_size=100)
results = s.run()
```

```python
all_data = dict(ids=[], links=[], names=[], parents=[], types=[], mirrors=[])
for keyword, d in results[0]._results.items():
    for k,v in d[0].items():
        all_data[k].extend(v)
```

```python
pd.DataFrame(data=all_data).drop_duplicates()
```
